```python
import pandas as pd
import os

```

```python
path = f"/Users/teresa/Desktop/dataverse_files/UN General Debate Corpus/UNGDC_1946-2022/"

sessions =os.listdir(path)
```

```python
def get_df_per_session(session_name):
    path = f"/Users/teresa/Desktop/dataverse_files/UN General Debate Corpus/UNGDC_1946-2022/{session_name}"
    files_in_folder = os.listdir(path)

    speeches = []
    for each in files_in_folder:
        print(each)
        with open(path+"/"+each) as file:
            lines =file.readlines()
            
        speeches.append(" ".join(lines))
        
    speakers =pd.read_excel("/Users/teresa/Desktop/dataverse_files/UN General Debate Corpus/Speakers_by_session.xlsx")

        
    speakers= speakers.astype({"Year":str})
    speakers["year_iso"] = speakers["Year"] + "_" + speakers["ISO Code"]
    speakers=speakers.reset_index()
    speakers["order"] =speakers.index
    list_of_iso = [each.split("_")[0] for each in files_in_folder]
    list_of_years = [each.split("_")[-1].split(".")[0] for each in files_in_folder]
    df_texts = pd.DataFrame({"iso":list_of_iso, "year":list_of_years})
    df_texts["session"] = 77

    df_texts["year_iso"] = df_texts["year"] + "_" + df_texts["iso"]

    df_texts["speeches"] = speeches
    merged_df = df_texts.merge(speakers, how="left", left_on=["year_iso"], right_on=["year_iso"])
    merged_df.sort_values("order", ascending=True)

    return merged_df
```

```python
session_list = []

for each in sessions:
    try:
        session_list.append(get_df_per_session(each))
    except:
        pass
```

```python
all_speeches = pd.concat(session_list)
all_speeches.fillna('NA', inplace=True)
```

```python
import csv
all_speeches.to_csv('all_speeches.csv', escapechar='\\')

```
